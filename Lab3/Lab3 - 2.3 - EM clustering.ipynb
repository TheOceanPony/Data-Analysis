# Lab 3 - Unsupervised learning

```python
import numpy as np
import matplotlib.pyplot as plt

from numpy.random import multivariate_normal
from sklearn.mixture import GaussianMixture

from time import time
```

---

## EM clustering

### Sample pool

```python
samplepool_size = 1000
means1, means2 = (8, 10), (2, 6)
X1 = multivariate_normal(means1, 5*np.identity(2), samplepool_size)
X2 = multivariate_normal(means2, 3*np.identity(2), samplepool_size)
X = np.concatenate([X1, X2])

labels = np.full(2*samplepool_size, 'r')
labels[:samplepool_size] = 'b'

plt.scatter(X[:,0], X[:,1], alpha=0.5, c=labels)
```

### Gaussian clustering

```python
clf = GaussianMixture(n_components=2, covariance_type='full')
clf.fit(X)
```

```python
print(f'PRED MEANS : {clf.means_[0]}, {clf.means_[1]}')
print('ACTUAL MEANS :',means2, means1)
```

```python
predicted = clf.predict(X)
color=['r','b']
for k in range(0,2):
    data = X[predicted==k]
    plt.scatter(data[:,0],data[:,1],c=color[k], alpha=0.5)
plt.axis('off')
plt.show()
```

```python
print('weight: \n',clf.weights_, '\n cov: \n', clf.covariances_)
```
